```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import datetime
```

```python
# not good method

events = [datetime.date(2015,1,23),
          datetime.date(2015,1,28),
          datetime.date(2015,2,3),
          datetime.date(2015,2,21),
          datetime.date(2015,3,15),
          datetime.date(2015,3,24),
          datetime.date(2015,4,8),
          datetime.date(2015,4,24)]

readings = [12,22,25,20,18,15,17,14]

plt.plot(events, readings)
plt.show()
```

```python
# this is good

months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y-%m')

events = [datetime.date(2015,1,23),
          datetime.date(2015,1,28),
          datetime.date(2015,2,3),
          datetime.date(2015,2,21),
          datetime.date(2015,3,15),
          datetime.date(2015,3,24),
          datetime.date(2015,4,8),
          datetime.date(2015,4,24)]

readings = [12,22,25,20,18,15,17,14]

fig, ax = plt.subplots(figsize=(12,8))

ax.plot(events, readings)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
```

```python
x1 = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
y1 = np.sin(3.1 * x) / x
y2 = np.sin(2 * x) / x
y3 = np.sin(3 * x) / x

plt.plot(x1, y1, color="b")
plt.plot(x1, y2, color='r')
plt.plot(x1, y3, color='g')

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-\pi$', r'$0$', r'$+\pi$', r'$+2\pi$'])
plt.yticks([-1,0,1,2,3],
           [r'$-1$', r'$0$', r'$+1$', r'$+2$', r'$+3$'])

plt.show()
```

```python
x1 = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
y1 = np.sin(3 * x) / x
y2 = np.sin(2 * x) / x
y3 = np.sin(x) / x

plt.plot(x1, y1, color="b")
plt.plot(x1, y2, color='r')
plt.plot(x1, y3, color='g')

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-\pi$', r'$0$', r'$+\pi$', r'$+2\pi$'])
plt.yticks([-1,0,1,2,3],
           [r'$-1$', r'$0$', r'$+1$', r'$+2$', r'$+3$'])

ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.show()
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.01)
y1 = np.sin(3 * x) / x
y2 = np.sin(2 * x) / x
y3 = np.sin(x) / x

plt.plot(x, y1, color="b")
plt.plot(x, y2, color="r")
plt.plot(x, y3, color="g")

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-\pi$', r'$0$', r'$+\pi$', r'$+2\pi$'])
plt.yticks([-1,0,+1,+2,+3], 
           [r'$-1$', r'$0$', r'$+1$', r'$+2$', r'$+3$'])

plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x}= 1$', 
             xy=[0,1],
             xycoords='data',
             xytext=[30,30],
             fontsize=16,
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->", connectionstyle='arc3, rad=.2'))

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['top'].set_position(('data', 0))

plt.show()
```
