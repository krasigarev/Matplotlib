```python
# bar charts

index = [0,1,2,3,4]
values = [5,7,3,4,6]

plt.bar(index, values)
plt.show()
```

```python
# bar chart

index = np.arange(5)

value_1 = [5,7,3,4,6]

plt.bar(index, value_1)
plt.xticks(index + 0.4, ['A', 'B', 'C', 'D', 'E'])

plt.show()
```

```python
index = np.arange(5)
value_2 = [5,7,3,4,6]
std_1 = [0.8, 1, 0.4, 0.9, 1.3]

plt.title("A bae Chart")
plt.bar(index, value_2, yerr=std_1, error_kw={'ecolor': '0.1', 'capsize': 6}, alpha=0.7, label='First')
plt.xticks(index + 0.4, ['A', 'B', 'C', 'D', 'E'])
plt.legend(loc=2)

plt.show()
```

```python
# Horizontal Bar Chart

index = np.arange(5)
value_3 = [5,7,3,4,6]
std_2 = [0.8, 1, 0.4, 0.9, 1.3]

plt.title("A Horizontal Bar Chart")
plt.barh(index, value_3, xerr=std_2, error_kw={'ecolor': '0.1', 'capsize':6}, alpha=0.7, label='First')
plt.yticks(index + 0.4, ["A", "B", "C", "D", "E"])
plt.legend(loc=2)

plt.show()
```

```python
#  Multiserial Bar Charts

index = np.arange(5)
y_1 = [5,7,3,4,6]
y_2 = [6,6,4,5,7]
y_3 = [5,6,5,4,6]

bw = 0.3 

plt.axis([0,5,0,8])

plt.title("A Multiseries Bar Chart", fontsize=20)
plt.bar(index, y_1, bw, color='b')
plt.bar(index + bw, y_2, bw, color='g')
plt.bar(index + 2 * bw, y_3, bw, color='r')
plt.xticks(index + 1.5 * bw, ['A', 'B', 'C', 'D', 'E'])

plt.show()
```

```python
series_1 = np.array([1,3,4,3])
series_2 = np.array([2,4,4,5])
series_3 = np.array([2,3,3,4])
index = np.arange(4)

plt.axis([-0.5,3.5,0,15])
plt.title("A Multiseries Stacked Bar Chart")

plt.bar(index, series_1, color='r')
plt.bar(index, series_2, color='b', bottom=series_1)
plt.bar(index, series_3, color='g', bottom=series_2 + series_1)
plt.xticks(index + 0.4, ['Jan18','Feb18','Mar18','Apr18'])

plt.show()
```

```python
series_1 = np.array([1,3,4,3])
series_2 = np.array([2,4,4,5])
series_3 = np.array([2,3,3,4])
index = np.arange(4)

plt.axis([0,15,-0.5,3.5])
plt.title("A Multiseries Horizontal Stacked Bar Chart")

plt.barh(index, series_1, color='r')
plt.barh(index, series_2, color='b', left=series_1)
plt.barh(index, series_3, color='g', left=(series_2 + series_1))
plt.yticks(index + 0.4, ['Jan18', 'Feb18','Mar18','Apr18'])

plt.show()
```

```python
series_1 = np.array([1,3,4,3])
series_2 = np.array([2,4,4,5])
series_3 = np.array([2,3,3,4])
index = np.arange(4)

plt.axis([0,15,-0.5,3.5])
plt.title('A Multiseries Horizontal Stacked Bar Chart')

plt.barh(index, series_1, color='w', hatch='xx')
plt.barh(index, series_2, left=series_1, color='w', hatch='///')
plt.barh(index, series_3, left=series_2 + series_1, color='w', hatch='\\\\\\')
plt.yticks(index + 0.4, ['Jan18','Feb18','Mar18','Apr18'])

plt.show()
```

```python
x_0 = np.arange(8)
y_1 = np.array([1,3,4,6,4,3,2,1])
y_2 = np.array([1,2,5,4,3,3,2,1])

plt.ylim(-7,7)
plt.bar(x_0, y_1, 0.9, facecolor='r')
plt.bar(x_0, -y_2, 0.9, facecolor='b')
plt.xticks(())
plt.grid(True)

for x, y in zip(x_0, y_1):
    plt.text(x, y, '%d' % y, ha='center', va='bottom')

for x, y in zip(x_0, y_2):
    plt.text(x, -y, '%d' % y, ha='center', va='top')

plt.show()
```

```python
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [10,30,45,15]
colors = ['yellow', 'green', 'red', 'blue']

explode = [0.3,0,0,0]

plt.title('A Pie Chart')
plt.pie(values, labels=labels, colors=colors, explode=explode, startangle=180)
plt.axis('equal')

plt.show()
```
