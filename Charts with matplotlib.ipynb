## Multiserial Bar Charts

```python
index = np.arange(5)
y_1 = [5,7,3,4,6]
y_2 = [6,6,4,5,7]
y_3 = [5,6,5,4,6]

bw = 0.3 

plt.axis([0,5,0,8])

plt.title("A Multiseries Bar Chart", fontsize=20)
plt.bar(index, y_1, bw, color='b')
plt.bar(index + bw, y_2, bw, color='g')
plt.bar(index + 2 * bw, y_3, bw, color='r')
plt.xticks(index + 1.5 * bw, ['A', 'B', 'C', 'D', 'E'])

plt.show()
```

## Stacked Bar Chart
- Multiseries Stacked Bar Chart

```python
series_1 = np.array([1,3,4,3])
series_2 = np.array([2,4,4,5])
series_3 = np.array([2,3,3,4])
index = np.arange(4)

plt.axis([-0.5,3.5,0,15])
plt.title("A Multiseries Stacked Bar Chart")

plt.bar(index, series_1, color='r')
plt.bar(index, series_2, color='b', bottom=series_1)
plt.bar(index, series_3, color='g', bottom=series_2 + series_1)
plt.xticks(index + 0.4, ['Jan18','Feb18','Mar18','Apr18'])

plt.show()
```

- Multiseries Horizontal Stacked Bar Chart

```python
series_1 = np.array([1,3,4,3])
series_2 = np.array([2,4,4,5])
series_3 = np.array([2,3,3,4])
index = np.arange(4)

plt.axis([0,15,-0.5,3.5])
plt.title("A Multiseries Horizontal Stacked Bar Chart")

plt.barh(index, series_1, color='r')
plt.barh(index, series_2, color='b', left=series_1)
plt.barh(index, series_3, color='g', left=(series_2 + series_1))
plt.yticks(index + 0.4, ['Jan18', 'Feb18','Mar18','Apr18'])

plt.show()
```

```python
series_1 = np.array([1,3,4,3])
series_2 = np.array([2,4,4,5])
series_3 = np.array([2,3,3,4])
index = np.arange(4)

plt.axis([0,15,-0.5,3.5])
plt.title('A Multiseries Horizontal Stacked Bar Chart')

plt.barh(index, series_1, color='w', hatch='xx')
plt.barh(index, series_2, left=series_1, color='w', hatch='///')
plt.barh(index, series_3, left=series_2 + series_1, color='w', hatch='\\\\\\')
plt.yticks(index + 0.4, ['Jan18','Feb18','Mar18','Apr18'])

plt.show()
```

## Contour Plots

```python
dx = 0.01
dy = 0.01

x = np.arange(start=-2.0, stop=2.0, step=dx)
y = np.arange(start=-2.0, stop=2.0, step=dy)

X,Y = np.meshgrid(x, y)

def f(x,y):
    return (1 - y ** 5 + x ** 5) * np.exp(-x ** 2 -y ** 2)

C = plt.contour(X, Y, f(X, Y), 8, colors='black')
plt.contourf(X, Y, f(X, Y), 8)
plt.clabel(C, inline=1, fontsize=10)

plt.show()
```

## Polar Chart

```python
N = 8

theta = np.arange(start=0.0, stop=2*np.pi, step=2*np.pi/N)
radii = np.array([4,7,5,3,1,5,6,7])

plt.axes([0.025, 0.025, 0.95, 0.95], polar=True)
color = np.array(['#4bb2c5', '#c5b47f', '#EAA228', '#579575', '#839557', '#958c12', '#953579', '#4b5de4'])
bars = plt.bar(theta, radii, width=(2*np.pi/N), bottom=0.0, color=colors)
plt.title("A Polar Chart")

plt.show()
```

## The mplot3d Toolkit

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

X = np.arange(start=-2, stop=2, step=0.1)
Y = np.arange(start=-2, stop=2, step=0.1)

X, Y = np.meshgrid(X, Y)

def f(x, y):
    return ( 1 - y ** 5 + x ** 5)* np.exp(-x**2 -y**2)

ax.plot_surface(X, Y, f(X, Y), rstride=1, cstride=1)

plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

X = np.arange(start=-2, stop=2, step=0.1)
Y = np.arange(start=-2, stop=2, step=0.1)

X, Y = np.meshgrid(X, Y)

def f(x, y):
    return ( 1 - y ** 5 + x ** 5)* np.exp(-x**2 -y**2)

ax.plot_surface(X, Y, f(X, Y), rstride=1, cstride=1, cmap=plt.cm.hot)
ax.view_init(elev=30, azim=125)
```

##  Scatter Plots in 3D

```python
xs = np.random.randint(30,40,100)
ys = np.random.randint(20,30,100)
zs = np.random.randint(10,20,100)
xs2 = np.random.randint(50,60,100)
ys2 = np.random.randint(30,40,100)
zs2 = np.random.randint(50,70,100)
xs3 = np.random.randint(10,30,100)
ys3 = np.random.randint(40,50,100)
zs3 = np.random.randint(40,50,100)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

ax.scatter(xs,ys,zs)
ax.scatter(xs2,ys2,zs2, c='r', marker='^')
ax.scatter(xs3,ys3,zs3, c='g', marker='*')
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")

plt.show()
```

## Bar Charts in 3D

```python
x = np.arange(8)
y = np.random.randint(0, 10, 8)
y1 = y + np.random.randint(0, 3, 8)
y2 = y1 + np.random.randint(0, 3, 8)
y3 = y2 + np.random.randint(0, 3, 8)
y4 = y3 + np.random.randint(0, 3, 8)

colors = ['#4bb2c5', '#c5b47f', '#EAA228', '#579575', '#839557', '#958c12', '#953579', '#4b5de4']

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')

ax.bar(x, y, 0, zdir='y', color=colors) 
ax.bar(x, y1, 10, zdir='y', color=colors)
ax.bar(x, y2, 20, zdir='y', color=colors) 
ax.bar(x, y3, 30, zdir='y', color=colors)
ax.bar(x, y4, 40,  zdir='y', color=colors)
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")
ax.view_init(elev=40)

plt.show()
```

## Grids of Subplots
